```python
import tweepy as tw
import pandas as pd
import networkx as nx
import json
```

```python
# Set up the authentication keys
with open('twitter_credentials.json', 'r') as file:
    creds = json.load(file)

# Twitter API key and API secret
my_api_key = creds['CONSUMER_KEY']
my_api_secret = creds['CONSUMER_SECRET']

# authenticate
auth = tw.AppAuthHandler(my_api_key, my_api_secret)
api = tw.API(auth, wait_on_rate_limit=True)

# Get my user ID number
me = api.get_user(screen_name = "@ggreenwald") #16076032
```

```python
# Fill a follower list with my followers (this is the fast bit)
user_list = ["16076032"]
follower_list = []
max_follower= 1000
for user in user_list:
    followers = []
    try:
        for page in tw.Cursor(api.get_follower_ids, user_id=user).pages():
            followers.extend(page)
            print(len(followers))
            if len(followers) > max_follower:
              break
    except tw.errors.TweepyException:
        print("error")
        continue
    follower_list.append(followers)

# Make a Dataframe for NetworkX

df = pd.DataFrame(columns=['source','target']).astype(str) # Empty DataFrame, astype() used throughout the script to prevent pandas from converting userIDs to scientific numbers
df['target'] = follower_list[0] # Set the list of followers as the target column
df['source'] = 16076032 # Set my user ID as the source 
df = df.astype(str)

# Here is the loop to retrieve followers of followers that takes ages

user_list = list(df['target']) #Use the list of followers collected in the previous step
for userID in user_list:
    print(userID)
    followers = []
    follower_list = []

    # fetching the user
    user = api.get_user(user_id = userID)

    # fetching the followers_count
    followers_count = user.followers_count

    try:
        for page in tw.Cursor(api.get_follower_ids, user_id=userID, count = 1000).pages():
            followers.extend(page)
            print(len(followers))
            if followers_count >= 1000: #Only take first 1000 followers to set some limits
                break
    except tw.errors.TweepyException:
        print("error")
        continue
    follower_list.append(followers)
    temp = pd.DataFrame(columns=['source', 'target']).astype(str)
    temp['target'] = follower_list[0]
    temp['source'] = userID
    df = df.append(temp).astype(str)
    df.to_csv("networkOfFollowers_ggreenwald.csv")

```
